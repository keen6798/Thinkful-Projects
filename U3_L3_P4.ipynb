```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Generate a normal distribution with mean=10 and standard deviation=5.
normal = np.random.normal(10, 5, 1000)

# Generate a binomial distribution with n=20 and p=0.5.
binomial = np.random.binomial(20, .5, 1000)

# Generate a uniform distribution with low=10 and high=20
uniform = np.random.uniform(0, 20, 1000)

# Generate a poisson distribution with lambda=2
poisson = np.random.poisson(1, 1000)

# Generate a exponential distribution with scale=1
exponential = np.random.exponential(1, 1000)

# Generate a chisquare distribution degrees of freedom=1
chisquare = np.random.chisquare(1, 1000)


distributions = pd.DataFrame()
distributions['vals'] = [normal,binomial,uniform,poisson,exponential,chisquare]
distributions.index = ['normal','binomial','uniform','poisson','exponential','chisquare']
```

```python
plt.figure(figsize=(15,10))

for i,j in enumerate(distributions.values):
    plt.subplot(2, 3, i+1)
    plt.hist(j, label='Mean = {} \n Std = {}'.format(j[0].mean(), j[0].std()))
    plt.axvline(j[0].mean(), color='c', linestyle='-', linewidth=2)
    plt.axvline(j[0].mean() + j[0].std(), color='c', linestyle='--', linewidth=2)
    plt.axvline(j[0].mean() - j[0].std(), color='c', linestyle='--', linewidth=2) 
    plt.title(distributions.index[i] + ' distribution')
    plt.legend(loc='center left', bbox_to_anchor=(0, -0.15))
    

plt.tight_layout()
plt.show()
```

$\textbf{Normal}:$ The mean is centralized which is a good estimate of the central tendency. Most of the data falls within one standard deviation.

$\textbf{Binomial}:$ I think that the data here is skewed. The mean is not quite centered, though it does seem that the most common values do fall within one standard deviation.

$\textbf{Uniform}:$ The mean is centered, but with a uniform distribution, there should not be any visible trend.

$\textbf{Poisson}:$ The mean is not a good estimate of central tendency and the data is skewed more towards 0.

$\textbf{Exponential}:$ The mean is not a good estimate of the central tendency and the data is skewed to towards 0. more extreme drop than poisson.

$\textbf{ChiSquare}:$ The mean is not a good estimate of the central tendency and the data is skewed to towards 0. More extreme drop than exponential. 

```python
rand1 = np.random.normal(5, .5, 100)
rand2 = np.random.normal(10, 1, 100)
rand3 = rand1 + rand2

plt.hist(rand3, label='Mean = {} \n Std = {}'.format(rand3.mean(), rand3.std()))
plt.axvline(rand3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(rand3.mean() + rand3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(rand3.mean() - rand3.std(), color='b', linestyle='dashed', linewidth=2) 
plt.legend(loc='center left', bbox_to_anchor=(0, -0.15))

plt.show()
```

We can see that for our new data set, the new mean is the two other means added together... 5 + 10 = 15. But the standard deviation is just a little greater than 1. That is because the new standard deviation is std3$=\sqrt{\text{std1}^2 + \text{std2}^2}$. The data is now showing some skew.
