```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set()
```

```python
planets = sns.load_dataset('planets')
planets.head()
```

# 1.  Choose one variable and plot that variable four different ways.

chosen variable: Method

```python
methods1 = sns.catplot('method', data=planets, kind='count', )
methods1.set_xticklabels(rotation=90);
```

This plot tells me how many exoplanets have been found using each one of detection methods. We can see here that the radial velocity detection method has detected the most amount of exoplanets.

```python
methods2 = sns.catplot('method', 'distance', data=planets, kind='box')
methods2.set_xticklabels(rotation=90);
```

This plot tells us information on how well each detection method detects exoplanets at certain distances. we can see that the microlensing method is far better at finding exoplanets at farther distances, but also doesn't do well at finding exoplanets up close either.

```python
methods3 = sns.catplot('method', 'number', data=planets, kind='swarm')
methods3.set_xticklabels(rotation=90);
```

This plot tells me about which method detects numerous planets or not. It looks like the transit method does well at detecting any number.

```python
methods4 = sns.catplot('method', 'orbital_period', data=planets, kind='bar')
methods4.set_xticklabels(rotation=90);
```

Here we can see that the imaging method is able to detecting a wider range, as well as exoplanets with larger orbital periods, usually indicating planets that are farther from their star.
